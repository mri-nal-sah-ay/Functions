Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list

```python
def sum_even_numbers():
    num_list = input("Enter a list of numbers: ").split()
    num_list = [int(num) for num in num_list]
    even_sum = sum(num for num in num_list if num % 2 == 0)
    return even_sum

print("Sum of even numbers:", sum_even_numbers())
```

Q2. Create a Python function that accepts a string and returns the reverse of that string

```python
def reverse_string():
    text = input("Enter String: ")
    return text[::-1]

print("Reversed string:", reverse_string())
```

Q3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number

```python
def square_numbers():
    num_list = input("Enter a list of integers: ").split()
    num_list = [int(num) for num in num_list]
    return [num ** 2 for num in num_list]

print("List of squares:", square_numbers())
```

Q4. Write a Python function that checks if a given number is prime or not from 1 to 200

```python
def is_prime():
    num = int(input("Enter a number between 1 and 200 to check if it's prime: "))
    if num < 1 or num > 200:
        return "Number must be between 1 and 200"
    if num == 1:
        return f"{num} is not a prime number"
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return f"{num} is not a prime number"
    return f"{num} is a prime number"

print(is_prime())
```

Q5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms

```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            self.count += 1
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            return result
        raise StopIteration

terms = int(input("Enter the number: "))
fib = FibonacciIterator(terms)
print("Fibonacci sequence:", [num for num in fib])
```

Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent

```python
def powers_of_two():
    max_exp = int(input("Enter the maximum exponent: "))
    for exp in range(max_exp + 1):
        yield 2 ** exp

print("Powers of 2:", list(powers_of_two()))
```

Q7.  Implement a generator function that reads a file line by line and yields each line as a string

```python
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

def read_file_lines():
    file_path = input("Enter the path to the file in Google Drive: ")
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        yield "File not found"
    except Exception as e:
        yield f"An error occurred: {str(e)}"

# Print lines from file
for line in read_file_lines():
    print(line)
```

Q8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple

```python
def sort_tuples():
    tuple_input = input("Enter tuples: ").split()
    tuple_list = [eval(t) for t in tuple_input]
    sorted_list = sorted(tuple_list, key=lambda x: x[1])
    return sorted_list

print("Sorted tuples:", sort_tuples())
```

Q9.  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

```python
def celsius_to_fahrenheit():
    celsius_list = input("Enter temperature: ").split()
    celsius_list = [float(temp) for temp in celsius_list]
    fahrenheit_list = list(map(lambda c: (c * 9/5) + 32, celsius_list))
    return fahrenheit_list

print("Temperatures in Fahrenheit:", celsius_to_fahrenheit())
```

Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string

```python
def remove_vowels():
    text = input("Enter a string: ")
    vowels = 'aeiouAEIOU'
    result = ''.join(filter(lambda char: char not in vowels, text))
    return result

print("String without vowels:", remove_vowels())
```
